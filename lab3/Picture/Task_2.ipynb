```python
import math as m
import pygame
from pygame.draw import *

# Developer params

# General
clock = pygame.time.Clock()
finished = False
f_p_s = 1

# Screen
screen_start_x = 0
screen_start_y = 0
screen_width = 1500
screen_hight = 800
screen = pygame.display.set_mode((screen_width, screen_hight))

# Background
background_color = [255, 222, 173]

# Panda
black = [0, 0, 0]
white = [255, 255, 255]
panda_1_x = 500
panda_1_y = 400
panda_1_size = 100
panda_2_x = 1100
panda_2_y = 400
panda_2_size = 60

# Tree
tree_1_x = 800
tree_2_x = 400
tree_3_x = 300
tree_4_x = 900
tree_1_y = 400
tree_2_y = 200
tree_3_y = 700
tree_4_y = 600
tree_1_hight = 100
tree_2_hight = 50
tree_3_hight = 20
tree_4_hight = 80
tree_1_width = 10
tree_2_width = 5
tree_3_width = 3
tree_4_width = 9
tree_color = [46, 139, 87] # green forest color

pygame.init()

def draw_tree(start_x:int, start_y:int, hight:int, width:int, tree_color:list):
    # start_x and start_y are coordinates of the bottom left point of the tree
    'Draws green tree'
    
    brunch_length_ratio = 20 # Ratio between length of brunch and width of tree
    decrease_ratio = 1.5 # Ratio between bottom segmet size and every diagonal upper one
    diagonal_segments_angle = m.pi / 6 # angle in rads
    diagonal_segments_amount = 2
    space = 10 # space between tree segments
    vertical_segments_amount = 2
    upper_brunch_size_multiplicator = 1.5
    upper_leaves_amount = 5
    bottom_leaves_amount = 3
    
    # Var simplification
    b_l_a = bottom_leaves_amount
    b_l_r = brunch_length_ratio
    d_s_a = diagonal_segments_angle
    h = hight
    w = width
    u_b_s_m = upper_brunch_size_multiplicator
    u_l_a = upper_leaves_amount
    
    
    for vertical_segment in range(vertical_segments_amount): # Draw vertical segments
        rect(screen, tree_color, (start_x, start_y - h, w, h))
        if vertical_segment % 2 == 1:
            for direction in range(2):
                draw_brunch(start_x, start_y, h, w * b_l_r, direction, tree_color, b_l_a)
        start_y = int(start_y - h * (1 + 1 / space))
    
    for diagonal_segment in range(diagonal_segments_amount): # Draw diagonal segments
        h = h / decrease_ratio
        space = space / decrease_ratio
        w = w / decrease_ratio
        polygon(screen, tree_color,[(start_x, start_y),
                                    (int(start_x + w * m.cos(d_s_a)), int(start_y + w * m.sin(d_s_a))),
                                    (int(start_x + h * m.sin(d_s_a) + w * m.cos(d_s_a)), int(start_y - h * m.cos(d_s_a) + w * m.sin(d_s_a))),
                                    (int(start_x + h * m.sin(d_s_a)), int(start_y - h * m.cos(d_s_a)))])
        if diagonal_segment % 2 == 1:
            for direction in range(2):
                brunch_scale = direction + 1 # Size of brunches decreaces with it number
                b_s = brunch_scale
                draw_brunch(start_x, start_y, u_b_s_m * h / b_s, u_b_s_m * w * b_l_r / b_s, direction, tree_color, u_l_a)
        start_x = int(start_x + h * m.sin(d_s_a) + space)
        start_y = int(start_y - h * m.cos(d_s_a) - space)


def draw_brunch(start_x:int, start_y:int, hight:int, width:int, direction:int, tree_color:list, leaves_amount:int):
    # Direction is 1 or 0 int. 1 is for left and 0 is for right
    'Draws brunch with leaves'
    
    brunch_leave_hight_ratio = 2 # Ratio between brunch and leave vertical size
    brunch_leave_width_ratio = 4 # Ratio between brunch and leave horizontal size
    start_angle = 0
    end_angle = m.pi
    leave_diff_ratio = 10 # Parameter to define difference in y_position of leaves
    leave_hight = int(hight / 2)
    leave_space = width / (2 * leaves_amount) # Distanse petween leaves and tree
    leave_start_y_ratio = 10 # Difference between brunch and leave y_position
    
    # Var simplification
    b_l_h_r = brunch_leave_hight_ratio
    b_l_w_r = brunch_leave_width_ratio
    d = direction
    x = start_x - width * d
    y = start_y - hight * d
    
    arc(screen, tree_color, (int(x), int(y), int(width), int(hight)), start_angle, end_angle)
    x = int(x + leave_space)
    y = int(y + hight / leave_start_y_ratio)
    
    for leave in range(leaves_amount): # Draw leaves
        ellipse(screen, tree_color, (x, y, int(width / (leaves_amount * b_l_w_r)), int(hight / b_l_h_r)))
        x = int(x + width / leaves_amount)
        y = int(y + hight / leave_diff_ratio * -1 * (-1)**leave)
        
        
def draw_panda(start_x:int, start_y:int, size:int, white:list, black:list):
    'Draws panda'
    
    body_head_ratio = 3
    head_y_x_ratio = 2
    
    # Var simplification
    b_h_r = body_head_ratio
    h_y_x_r = head_y_x_ratio
    
    # Draw body
    ellipse(screen, white, (start_x, start_y,  size * b_h_r, size * b_h_r // h_y_x_r))
    
    # Draw forward right leg
    ellipse(screen, black, (start_x, start_y,  size // 2, size * 2))
    polygon(screen, black, [(start_x + size // 2, start_y),
                            (start_x + size, start_y + size),
                            (start_x + size // 4, start_y + size * 2)])
    
    # Draw forward left leg
    polygon(screen, black, [(start_x + 3 * size // 2, start_y),
                            (start_x + 3 * size // 2, start_y + size),
                            (start_x + 13 * size // 12, start_y + 9 * size // 5),
                            (start_x + 2 * size // 3, start_y + 8 *size // 5)])
    circle(screen, black, (start_x + 5 * size // 6, start_y + 9 *size // 5), size // 4)
    
    # Draw backward left leg
    polygon(screen, black, [(start_x + 5 * size // 2, start_y + size // 2),
                            (start_x + 5 * size // 2, start_y + 3 * size // 2),
                            (start_x + 22 * size // 12, start_y + 2 * size),
                            (start_x + 13 * size // 8, start_y + 2 * size)])
    ellipse(screen, black, (start_x + 13 * size // 8, start_y + 31 * size // 16, size // 4, size // 8))
    ellipse(screen, black, (start_x + 19 * size // 8, start_y + size // 2, size // 4, size))
    
    # Draw head
    ellipse(screen, white, (start_x, start_y, size // h_y_x_r, size))
    ellipse(screen, white, (start_x + size, start_y, size // h_y_x_r, size))
    circle(screen, white, (start_x + size // h_y_x_r, start_y + size), size // 3)
    polygon(screen, white, [(start_x + size // (2 * h_y_x_r), start_y),
                            (start_x + size + size // (2 * h_y_x_r), start_y),
                            (start_x + size + size // (2 * h_y_x_r), start_y + size),
                            (start_x + size // h_y_x_r, start_y + size + size // 3)])
    ellipse(screen, white, (start_x + size // (2 * h_y_x_r), start_y - size // (2 * h_y_x_r), size, size // h_y_x_r))
    
    # Draw left ear
    circle(screen, black, (start_x + size // (2 * h_y_x_r), start_y - size // (3 * h_y_x_r)), size // 5)
    circle(screen, black, (start_x + size // h_y_x_r, start_y - size // (2 * h_y_x_r)), size // 7)
    
    # Draw right ear
    circle(screen, black, (start_x + size + size // (2 * h_y_x_r), start_y - size // (3 * h_y_x_r)), size // 5)
    circle(screen, black, (start_x + size // 2 + size // h_y_x_r, start_y - size // (2 * h_y_x_r)), size // 7)
    
    # Draw left eye
    ellipse(screen, black, (start_x, start_y + size // 3, size // (3 * h_y_x_r), size // 3))
    
    # Draw right eye
    circle(screen, black, (start_x + size // 2, start_y + size // 2), size // 6)
    
    # Draw nose
    ellipse(screen, black, (start_x, start_y + (4 * size // 5),  size // 3, size // (2 * h_y_x_r)))
    
rect(screen, background_color, (screen_start_x, screen_start_y, screen_width, screen_hight)) # Draw background
draw_tree(tree_1_x, tree_1_y, tree_1_hight, tree_1_width, tree_color)
draw_tree(tree_2_x, tree_2_y, tree_2_hight, tree_2_width, tree_color)
draw_tree(tree_3_x, tree_3_y, tree_3_hight, tree_3_width, tree_color)
draw_tree(tree_4_x, tree_4_y, tree_4_hight, tree_4_width, tree_color)
draw_panda(panda_1_x, panda_1_y, panda_1_size, white, black)
draw_panda(panda_2_x, panda_2_y, panda_2_size, white, black)

pygame.display.update()
while not finished:
    clock.tick(f_p_s)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True

pygame.quit()
```
